In this exercise, we will calculates a derivative. For that purpose, keep in mind the mathematical definition of a derivative is

 dfdx=limδx→0f(x+δx)−f(x)δx

Write a script that calculate the derivative of f(x)=x2 on 101 points between 0 and 10 for δx=10−2. 
Plot the error curve (i.e. expected value – computed value) over the 101 points.
Do the same for different values of different values on δx ranging from 10−2 to 10−14. How does the error vary with the δx?

```

# calculating the derrivative for a single delta_x
import numpy as np
import matplotlib.pyplot as plt


def der(func, x, delta_x):
    return (func(x + delta_x) - func(x))/delta_x


def func(x):
    return x**2


delta_x = 0.01
x_values = np.linspace(0, 10, 101)

computed_values = [der(func, x, delta_x) for x in x_values]
expected_values = [2*x for x in x_values]
differences = [expected_values[i] - computed_values[i] for i in range(len(x_values))]

plt.plot(x_values, differences)

```

```
# derivative for varying delta_x

# deltas = np.linspace(10**-2, 10**-18, 50)
deltas = [10**-2, 10**-3, 10**-4, 10**-5, 10**-6, 10**-7, 10**-8, 10**-9, 10**-10, 10**-11, 10**-12,
          10**-13, 10**-14, 10**-15, 10**-16, 10**-14, 10**-18, 10**-19, 10**-20]

mean_error = []
for i in range(len(deltas)):
    delta = deltas[i]
    # delta = 1.00000000e-15
    computed_values = [der(func, x, delta) for x in x_values]
    differences = [expected_values[x] - computed_values[x] for x in range(len(x_values))]
    mean_error.append(np.mean(np.abs(differences)))
# print(mean_error)
plt.plot(deltas, mean_error)

mean_error = []
delta = 1.11000000e-14
computed_values = [der(func, x, delta) for x in x_values]
differences = [expected_values[x] - computed_values[x] for x in range(len(x_values))]
mean_error.append(np.mean(np.abs(differences)))
print(mean_error)
```
