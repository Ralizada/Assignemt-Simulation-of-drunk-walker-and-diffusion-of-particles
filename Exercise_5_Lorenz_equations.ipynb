Lorenz equations are a set of three coupled differential that may exhibit chaotic behavior. This model correspond to a simplification of a model used to predict meteorological phenomenon. The three coupled equations are:

dxdt=σ⋅(y−x)

dydt=ρ⋅x−y−x⋅z

dzdt=x⋅y−β⋅z

where x, y and z are the variables and σ , ρ and β are model parameters.

Write a program that solves Lorenz equations for σ=10, ρ=28 and β=3. Use the time range [0;50] with at least 1000 points.
Plot x, y and z as a function of time for different initial conditions. Which are the requirements on initial conditions to obtain chaotic behavior?
Plot x versus y phase diagram when the behavior is chaotic. Which shape do you obtain?
Play with the σ, ρ and beta.  Which are the requirements on the parameters to obtain chaotic behavior?

```
# sigma=10, ro=28 and beta=3
# dx/dt = sigma*(y - x)
# dy/dt = ro*x - y - x*z
# dz/dt = x*y - beta*z

import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Note: t0 is required for the odeint function, though it's not used here.


def lorentz_deriv(params, t0, sigma, beta, rho):
    x, y, z = params
    """Compute the time-derivative of a Lorenz system."""
    return [sigma * (y - x), x * (rho - z) - y, x * y - beta * z]


# initial conditions
# changes on x,y or both causes chaotic behaviour, not on z
x0 = [1, 0, 0]

# ro>2
sigma, beta, ro = [10, 3, 10]

t = np.linspace(0, 50, 100000)
x_t = integrate.odeint(lorentz_deriv, x0, t, args=(sigma, beta, ro))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_t[:, 0], x_t[:, 1], x_t[:, 2])

f, ax = plt.subplots(3, sharex=True)
for i in range(3):
    ax[i].plot(t, x_t[:, i])
plt.show()
# plt.plot(x_t[:, 0], x_t[:, 1])


```
